```python
import re

def initialize():
    print("Greetings!!!...Welcome to the portal.")
    
def email_check(mail):    
  mail_format="^[A-Za-z]+[A-Za-z|0-9._]+@[A-Za-z]+\.[A-Z|a-z]{2,}"
  Symbols = ['.', '!', '@', '#', '$', '%', '^', '&', '*']
  if re.search(mail_format,mail):
    return mail
  else:
    print(" \n Invalid Email ID \n")
    print(" \n The Email id should be in particular format:  \"[Sherlock@gmail.com]\" \n")
    return False

def password_check(pswd):
  pattern=re.compile(r'')
  if len(pswd)<5 or len(pswd)>16:
    print("Password length must be in range between 6 to 15 characters \n")
  elif re.search(r'[!@#$%&]',pswd) is None:
    print("Password must have atleast one Special Character \n")
  elif re.search(r'[0-9]',pswd) is None:
    print("Password must have atleast one numerical \n")
  elif re.search(r'[A-Z]',pswd) is None:
    print("Password must have atleast one Uppercase letter \n")
  elif re.search(r'[a-z]',pswd) is None:
    print("Password must have atleast one Lowercase letter \n")
  elif re.match(r'[a-z A-Z 0-9 !@#$%&]{6,15}',pswd):
    print("Valid Password")
    return True
  else:
    print("Invalid Password")
    return False

def user_credentials():
  mail=input("Enter the Email ID:", end=" ")
  mail_1=email_check(mail)
  D1=0
  if email_check(mail_1) == False:
    print(" \n Please enter valid Email ID")
    D1=D1+1
  else:
    pswd=input("Enter Password:", end=" ")
    if password_check(pswd) == False:
      print("\n Please enter valid password")
      D1=D1+1

  if D1==0:
    db = open("Database.txt","r")
    E=[mail,pswd]
    db.writelines(E)
    print("Email ID and Password are successfully registered \n")  

def forgot_password(ab):
  mail=user_credentials()
  index=list(ab.Email).index(mail)
  real_paswd=ab.Password[index]
  print("Your Password is: ",real_paswd)

def signup(ab):
  mail=user_credentials()
  if mail in ab.mail.info:
    print("Your Email ID is already registered. Try 1 for Login")
  login(ab)

def login(ab):
  mail=user_credentials()
  if mail not in ab.mail.info:
      print("Not Registered, Try 2 for Signup")
      signup(ab)
  elif mail in ab.mail.info:
    pwd=user_credentials()
    index=list(ab.Email).index(mail)
    real_paswd=ab.Password[index]
    while pwd!=real_paswd:
      print("Wrong Password Try Again")
      pwd=user_credentials()
    print("Access Granted")

def main(ab):
  initialize()
  a=input("Select the option:\n Enter 1 for Login \n  Enter 2 for Signup \n Enter 3 for Forgot Password")
  if a=='1':
    login(ab)
    return True
  elif a=='2':
    signup(ab)
    return True
  elif a=='3':
    forgot_password(ab)
    return True
  else:
    print("Give a Valid input")
    return False  
```
